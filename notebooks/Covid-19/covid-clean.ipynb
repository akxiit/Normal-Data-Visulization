What are the total number of rows and columns in the dataset?

Extract all rows where the WHO Region is "Europe."

```python
df[df['WHO Region'] == 'Europe']
```

Find all records where the number of Confirmed cases is greater than 10,000.

```python
df[df['Confirmed'] > 10000]
```

Sort the dataset by Confirmed cases in descending order.

```python
df.sort_values(by='Confirmed', ascending=False)
```

Calculate the total number of Deaths for each WHO Region.

```python
df.groupby('WHO Region')['Deaths'].sum()
```

Determine the country with the highest total Recovered cases.

```python
df.loc[df['Recovered'].idxmax()]['Country/Region']
```

Group the data by Date and calculate the total Confirmed cases for each day.

```python
df.groupby('Date')['Confirmed'].sum()
```

Find the average Active cases for each WHO Region.

```python
df.groupby('WHO Region')['Active'].mean()
```

Identify how many null values exist in the Province/State column.

```python
df['Province/State'].isna().count()
```

Replace null values in the Province/State column with "Unknown".

```python
df['Province/State'].fillna('Unknown', inplace=True)
```

Convert the Date column to datetime format and find the earliest and latest dates in the dataset.

Create a NumPy array of the Confirmed cases and calculate the mean, median, and standard deviation.

```python
arr = np.array(df['Confirmed'])
arr.mean() , arr.std() , np.median(arr)
```

Plot a line graph showing the total Confirmed cases over time.

```python
plt.plot('Confirmed' , 'Date' , data=df)   
```

Create a bar chart showing the total Deaths for the top 5 countries.

```python
country = df.groupby('Country/Region')['Deaths'].sum().sort_values(ascending=False).head(5)

plt.bar(country.index , country.values)
```

Use Seaborn to create a heatmap showing the correlation between Confirmed, Deaths, Recovered, and Active cases.

```python
corr_matrix = df[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

Create a scatter plot to visualize the relationship between Confirmed and Deaths.

```python
plt.scatter('Confirmed', 'Deaths', data=df)
```
